```python
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

def dataset_to_coco_json(dataset, dataset_name="student_id_dataset"):
    """
    Convert a custom dataset into a COCO-style annotation format.

    This function takes a custom Dataset object and transforms it into a dictionary
    that follows the COCO annotation format. The COCO format requires three main keys:
        1) "images":      each image entry should have {"id", "file_name", "width", "height"}
        2) "annotations": each annotation entry should have 
                          {"id", "image_id", "category_id", "bbox", "area", "iscrowd"}
        3) "categories":  each category entry should have {"id", "name"}

    - BBox Format in COCO:
      The bounding boxes must be in the format [x_min, y_min, width, height] rather than 
      the [x1, y1, x2, y2] format. This function converts from one format to the other.

    - Skipping the "background" Class:
      The background class (if it exists) typically does not appear as an annotated category 
      in object detection tasks, so it is excluded from the COCO output.

    Args:
        dataset (torch.utils.data.Dataset):
            A Dataset that, when indexed (e.g., dataset[i]), 
            returns a tuple (image, target_dict).
            The target_dict must contain at least:
              target_dict['boxes']:   bounding boxes in [x1, y1, x2, y2] format
              target_dict['labels']:  class labels (integers)
              target_dict['masks']:   segmentation masks (optional)
        dataset_name (str):
            A descriptive name for your dataset. 
            This will appear in the "info" field of the COCO JSON.

    Returns:
        tuple:
            coco_dict (dict):
                A dictionary containing the keys "info", "images", "annotations", and "categories".
                This dictionary can be saved to a JSON file for COCO evaluation tools.
            img_id_to_filename (dict):
                A helper dictionary mapping each image's unique COCO ID to its file name.
                Useful for debugging and ensuring correct image-to-annotation matching.
    """
    # In a typical COCO dataset, the first 'categories' entry is 'background' if you have it, 
    # but we skip it in "categories" because COCO usually doesn't track background as an object.
    # 
    # Example of the final structure:
    # coco_dict = {
    #   "info":       {...},
    #   "images":     [{"id", "file_name", "width", "height"}, ...],
    #   "annotations":[{"id", "image_id", "category_id", "bbox", "area", "iscrowd"}, ...],
    #   "categories": [{"id", "name"}, ...]
    # }

    coco_images = []
    coco_annotations = []
    coco_categories = []

    # 'class_names' should be defined or accessible within your code context:
    # For example, class_names = ["background", "text", "photo_id", ...]
    # Start building the categories. The ID (cat_id) should match how your
    # dataset encodes 'labels' for each class. Typically, label=0 = background.
    for idx, class_name in enumerate(class_names):
        if class_name == "background":
            continue  # skip background as an annotated category
        cat_id = idx  # or another numeric ID scheme, but must align with your dataset
        coco_categories.append({"id": cat_id, "name": class_name})
    
    # annotation ID for each object in the dataset must be unique
    ann_id_counter = 1

    # We iterate over every (image, target) in the dataset
    for img_index in range(len(dataset)):
        img, target = dataset[img_index]

        # Retrieve the width and height from the transformed image shape:
        # shape is typically (channels, height, width)
        height, width = img.shape[-2], img.shape[-1]
        
        # COCO uses a 1-based index for images
        image_id = img_index + 1
        
        # Suppose the dataset keeps a list of image filenames in `dataset._img_keys`
        # Adjust as needed for your code
        file_name = dataset._img_keys[img_index]

        # Store this image information in the "images" field
        coco_images.append({
            "id": image_id,
            "file_name": file_name,
            "width": width,
            "height": height
        })

        # Extract bounding boxes, labels, and possibly segmentation masks
        boxes = target['boxes']
        labels = target['labels']
        masks = target.get('masks', None)  # Only used if segmentation is relevant

        # Convert each bounding box from [x1, y1, x2, y2] to [x_min, y_min, w, h]
        for box_idx in range(boxes.shape[0]):
            x1, y1, x2, y2 = boxes[box_idx].tolist()
            w = x2 - x1
            h = y2 - y1
            area = w * h

            class_id = labels[box_idx].item()
            if class_id == 0:
                # If label=0 is background, skip it from the final annotations
                continue

            # Build the annotation entry
            ann_dict = {
                "id": ann_id_counter,
                "image_id": image_id,
                "category_id": class_id,
                "bbox": [x1, y1, w, h],
                "area": area,
                "iscrowd": 0  # By default, we consider each instance as a separate object
            }
            coco_annotations.append(ann_dict)
            ann_id_counter += 1

    # Build the final COCO dictionary with the required fields
    coco_dict = {
        "info": {
            "description": dataset_name,
            "year": 2023
        },
        "images": coco_images,
        "annotations": coco_annotations,
        "categories": coco_categories
    }

    # Create a helper map from image_id to image file name
    img_id_to_filename = {im["id"]: im["file_name"] for im in coco_images}

    return coco_dict, img_id_to_filename


def get_model_predictions_in_coco_format(model, dataloader, device, score_threshold=0.05):
    """
    Generate model predictions in COCO detection format.

    This function runs a trained detection model (e.g., Mask R-CNN or Faster R-CNN)
    over a DataLoader, then processes the output into a Python list of dictionaries
    matching the COCO detection results specification. Each dictionary in the returned
    list represents a single object detection prediction.

    COCO Detection Format Example:
        [
          {
            "image_id": int,
            "category_id": int,
            "bbox": [x,y,width,height],
            "score": float
          },
          ...
        ]

    Args:
        model (nn.Module):
            A PyTorch model for object detection, typically in evaluation mode.
        dataloader (torch.utils.data.DataLoader):
            A DataLoader that yields batches of images (and, optionally, targets).
        device (torch.device or str):
            The device ("cpu" or "cuda") on which the computations will be performed.
        score_threshold (float):
            Minimum confidence score to keep a predicted bounding box. 
            Predictions below this threshold will be filtered out.

    Returns:
        results_coco (list of dict):
            Each dict has the keys: "image_id", "category_id", "bbox", and "score".
            This format is exactly what pycocotools needs to evaluate detection predictions.
    """
    model.eval()
    results_coco = []

    # This offset aligns image indices with the 1-based "image_id" in COCO
    img_id_offset = 1

    # Use tqdm to get a nice progress bar while collecting predictions
    for idx, (images, targets) in enumerate(tqdm(dataloader, desc="Collecting predictions")):
        # In many custom DataLoaders, 'images' is a tuple/list of Tensors
        # Here we stack them into a single batch to run inference in one go
        images = torch.stack(images).to(device)

        with torch.no_grad():
            outputs = model(images)

        # Each element in 'outputs' corresponds to one image in the batch
        for batch_i, output in enumerate(outputs):
            # Determine the image_id to match how it was defined in dataset_to_coco_json
            image_id = (idx * dataloader.batch_size) + batch_i + img_id_offset

            # Extract predicted boxes, labels, and scores
            boxes = output['boxes'].cpu().numpy()
            scores = output['scores'].cpu().numpy()
            labels = output['labels'].cpu().numpy()

            # Filter out predictions below the chosen confidence threshold
            keep = scores >= score_threshold
            boxes = boxes[keep]
            scores = scores[keep]
            labels = labels[keep]

            # Convert bounding boxes from [x1, y1, x2, y2] to [x_min, y_min, w, h]
            for box, score, label in zip(boxes, scores, labels):
                x1, y1, x2, y2 = box
                w = x2 - x1
                h = y2 - y1

                # Build a single prediction entry
                coco_pred = {
                    "image_id": image_id,
                    "category_id": int(label),  # must match your dataset's category IDs
                    "bbox": [float(x1), float(y1), float(w), float(h)],
                    "score": float(score)
                }
                results_coco.append(coco_pred)

    return results_coco


def coco_evaluate(coco_gt_dict, coco_predictions):
    """
    Evaluate predictions using the COCO evaluation metrics.

    This function uses pycocotools to evaluate object detection predictions. It 
    constructs a COCO object from your ground-truth annotations (in-memory), loads 
    the detection results, and then computes standard COCO metrics (e.g., AP, AR).

    Args:
        coco_gt_dict (dict):
            The ground-truth COCO-style dictionary. 
            Usually returned by `dataset_to_coco_json()`.
        coco_predictions (list of dict):
            The model predictions in COCO detection format. 
            Usually returned by `get_model_predictions_in_coco_format()`.

    Returns:
        metrics_map (dict):
            A dictionary of common COCO metrics such as:
            - AP (IoU=0.50:0.95)
            - AP (IoU=0.50)
            - AP (IoU=0.75)
            - AP (small, medium, large)
            - AR@1, AR@10, AR@100 (and size-based AR).
            Each entry in the dict corresponds to a particular COCO metric.
    """
    # Create a blank COCO object and load your ground-truth annotations
    coco_gt = COCO()
    coco_gt.dataset = coco_gt_dict
    coco_gt.createIndex()

    # Load your detection predictions into COCO for evaluation
    coco_dt = coco_gt.loadRes(coco_predictions)

    # Run COCO evaluation. 'iouType' can be 'bbox' or 'segm' based on your data
    coco_eval = COCOeval(coco_gt, coco_dt, iouType='bbox')
    coco_eval.evaluate()
    coco_eval.accumulate()
    coco_eval.summarize()

    # Common metrics returned in coco_eval.stats (array of length 12):
    #  [0] mAP@IoU=0.50:0.95
    #  [1] mAP@IoU=0.50
    #  [2] mAP@IoU=0.75
    #  [3] mAP (small)
    #  [4] mAP (medium)
    #  [5] mAP (large)
    #  [6] AR@1
    #  [7] AR@10
    #  [8] AR@100
    #  [9] AR@100 (small)
    # [10] AR@100 (medium)
    # [11] AR@100 (large)

    metrics_map = {
        "AP (IoU=0.50:0.95)": coco_eval.stats[0],
        "AP (IoU=0.50)":      coco_eval.stats[1],
        "AP (IoU=0.75)":      coco_eval.stats[2],
        "AP (small)":         coco_eval.stats[3],
        "AP (medium)":        coco_eval.stats[4],
        "AP (large)":         coco_eval.stats[5],
        "AR@1":               coco_eval.stats[6],
        "AR@10":              coco_eval.stats[7],
        "AR@100":             coco_eval.stats[8],
        "AR@100 (small)":     coco_eval.stats[9],
        "AR@100 (medium)":    coco_eval.stats[10],
        "AR@100 (large)":     coco_eval.stats[11],
    }

    return metrics_map
```

#### Define a function to convert segmentation polygons to images

```python
def create_polygon_mask(image_size, vertices):
    """
    Create a grayscale image with a white polygonal area on a black background.

    Parameters:
    - image_size (tuple): A tuple representing the dimensions (width, height) of the image.
    - vertices (list): A list of tuples, each containing the x, y coordinates of a vertex
                        of the polygon. Vertices should be in clockwise or counter-clockwise order.

    Returns:
    - PIL.Image.Image: A PIL Image object containing the polygonal mask.
    """

    # Create a new black image with the given dimensions
    mask_img = Image.new('L', image_size, 0)
    
    # Draw the polygon on the image. The area inside the polygon will be white (255).
    ImageDraw.Draw(mask_img, 'L').polygon(vertices, fill=(255))

    # Return the image with the drawn polygon
    return mask_img
```

### Training Dataset Class

```python
class StudentIDDataset(Dataset):
    """
    This class represents a PyTorch Dataset for a collection of images and their annotations.
    The class is designed to load images along with their corresponding segmentation masks, bounding box annotations, and labels.
    """
    def __init__(self, img_keys, annotation_df, img_dict, class_to_idx, transforms=None):
        """
        Constructor for the HagridDataset class.

        Parameters:
        img_keys (list): List of unique identifiers for images.
        annotation_df (DataFrame): DataFrame containing the image annotations.
        img_dict (dict): Dictionary mapping image identifiers to image file paths.
        class_to_idx (dict): Dictionary mapping class labels to indices.
        transforms (callable, optional): Optional transform to be applied on a sample.
        """
        super(Dataset, self).__init__()
        
        self._img_keys = img_keys  # List of image keys
        self._annotation_df = annotation_df  # DataFrame containing annotations
        self._img_dict = img_dict  # Dictionary mapping image keys to image paths
        self._class_to_idx = class_to_idx  # Dictionary mapping class names to class indices
        self._transforms = transforms  # Image transforms to be applied
        
    def __len__(self):
        """
        Returns the length of the dataset.

        Returns:
        int: The number of items in the dataset.
        """
        return len(self._img_keys)
        
    def __getitem__(self, index):
        """
        Fetch an item from the dataset at the specified index.

        Parameters:
        index (int): Index of the item to fetch from the dataset.

        Returns:
        tuple: A tuple containing the image and its associated target (annotations).
        """
        # Retrieve the key for the image at the specified index
        img_key = self._img_keys[index]
        # Get the annotations for this image
        annotation = self._annotation_df.loc[img_key]
        # Load the image and its target (segmentation masks, bounding boxes and labels)
        image, target = self._load_image_and_target(annotation)
        
        # Apply the transformations, if any
        if self._transforms:
            image, target = self._transforms(image, target)
        
        return image, target

    def _load_image_and_target(self, annotation):
        """
        Load an image and its target (bounding boxes and labels).

        Parameters:
        annotation (pandas.Series): The annotations for an image.

        Returns:
        tuple: A tuple containing the image and a dictionary with 'boxes' and 'labels' keys.
        """
        # Retrieve the file path of the image
        filepath = self._img_dict[annotation.name]
        # Open the image file and convert it to RGB
        image = Image.open(filepath).convert('RGB')
        
        # Convert the class labels to indices
        labels = [shape['label'] for shape in annotation['shapes']]
        labels = torch.Tensor([self._class_to_idx[label] for label in labels])
        labels = labels.to(dtype=torch.int64)

        # Convert polygons to mask images
        shape_points = [shape['points'] for shape in annotation['shapes']]
        xy_coords = [[tuple(p) for p in points] for points in shape_points]
        mask_imgs = [create_polygon_mask(image.size, xy) for xy in xy_coords]
        masks = Mask(torch.concat([Mask(transforms.PILToTensor()(mask_img), dtype=torch.bool) for mask_img in mask_imgs]))

        # Generate bounding box annotations from segmentation masks
        bboxes = BoundingBoxes(data=torchvision.ops.masks_to_boxes(masks), format='xyxy', canvas_size=image.size[::-1])
                
        return image, {'masks': masks,'boxes': bboxes, 'labels': labels}
```

### Image Transforms

```python
# Compose transforms for data augmentation
data_aug_tfms = transforms.Compose(
    transforms=[
        iou_crop,
        transforms.ColorJitter(
                brightness = (0.875, 1.125),
                contrast = (0.5, 1.5),
                saturation = (0.5, 1.5),
                hue = (-0.05, 0.05),
        ),
        transforms.RandomGrayscale(),
        transforms.RandomEqualize(),
        transforms.RandomPosterize(bits=3, p=0.5),
        transforms.RandomHorizontalFlip(p=0.5),
    ],
)

# Compose transforms to resize and pad input images
resize_pad_tfm = transforms.Compose([
    resize_max, 
    pad_square,
    transforms.Resize([train_sz] * 2, antialias=True)
])

# Compose transforms to sanitize bounding boxes and normalize input data
final_tfms = transforms.Compose([
    transforms.ToImage(), 
    transforms.ToDtype(torch.float32, scale=True),
    transforms.SanitizeBoundingBoxes(),
])

# Define the transformations for training and validation datasets
train_tfms = transforms.Compose([
    data_aug_tfms, 
    resize_pad_tfm, 
    final_tfms
])
valid_tfms = transforms.Compose([resize_pad_tfm, final_tfms])
```

## Fine-tuning the Model

### Define the Training Loop

```python
def run_epoch(model, dataloader, optimizer, lr_scheduler, device, scaler, epoch_id, is_training):
    """
    Function to run a single training or evaluation epoch.
    
    Args:
        model: A PyTorch model to train or evaluate.
        dataloader: A PyTorch DataLoader providing the data.
        optimizer: The optimizer to use for training the model.
        loss_func: The loss function used for training.
        device: The device (CPU or GPU) to run the model on.
        scaler: Gradient scaler for mixed-precision training.
        is_training: Boolean flag indicating whether the model is in training or evaluation mode.
    
    Returns:
        The average loss for the epoch.
    """
    # Set the model to training mode
    model.train()
    
    epoch_loss = 0  # Initialize the total loss for this epoch
    progress_bar = tqdm(total=len(dataloader), desc="Train" if is_training else "Eval")  # Initialize a progress bar
    
    # Loop over the data
    for batch_id, (inputs, targets) in enumerate(dataloader):
        # Move inputs and targets to the specified device
        inputs = torch.stack(inputs).to(device)
        
        # Forward pass with Automatic Mixed Precision (AMP) context manager
        with autocast(torch.device(device).type):
            if is_training:
                losses = model(inputs.to(device), move_data_to_device(targets, device))
            else:
                with torch.no_grad():
                    losses = model(inputs.to(device), move_data_to_device(targets, device))
        
            # Compute the loss
            loss = sum([loss for loss in losses.values()])  # Sum up the losses

        # If in training mode, backpropagate the error and update the weights
        if is_training:
            if scaler:
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                old_scaler = scaler.get_scale()
                scaler.update()
                new_scaler = scaler.get_scale()
                if new_scaler >= old_scaler:
                    lr_scheduler.step()
            else:
                loss.backward()
                optimizer.step()
                lr_scheduler.step()
                
            optimizer.zero_grad()

        # Update the total loss
        loss_item = loss.item()
        epoch_loss += loss_item
        
        # Update the progress bar
        progress_bar_dict = dict(loss=loss_item, avg_loss=epoch_loss/(batch_id+1))
        if is_training:
            progress_bar_dict.update(lr=lr_scheduler.get_last_lr()[0])
        progress_bar.set_postfix(progress_bar_dict)
        progress_bar.update()

        # If loss is NaN or infinity, stop training
        if is_training:
            stop_training_message = f"Loss is NaN or infinite at epoch {epoch_id}, batch {batch_id}. Stopping training."
            assert not math.isnan(loss_item) and math.isfinite(loss_item), stop_training_message

    # Cleanup and close the progress bar 
    progress_bar.close()
    
    # Return the average loss for this epoch
    return epoch_loss / (batch_id + 1)

def train_loop(model, 
               train_dataloader, 
               valid_dataloader, 
               optimizer,  
               lr_scheduler, 
               device, 
               epochs, 
               checkpoint_path, 
               use_scaler=False):
    """
    Main training loop.
    """
    scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' and use_scaler else None
    best_loss = float('inf')

    # ### NEW or UPDATED ###
    # Build the GT COCO JSON once outside the loop if your dataset is unchanging
    val_coco_gt_dict, _ = dataset_to_coco_json(valid_dataset, dataset_name="student_id_dataset")

    for epoch in tqdm(range(epochs), desc="Epochs"):
        # --------------------
        # 1) Training
        # --------------------
        train_loss = run_epoch(model, train_dataloader, optimizer, lr_scheduler, device, scaler, epoch, is_training=True)
        
        # --------------------
        # 2) Validation
        # --------------------
        with torch.no_grad():
            valid_loss = run_epoch(model, valid_dataloader, None, None, device, scaler, epoch, is_training=False)
        
        # --------------------
        # 3) COCO Evaluation
        # --------------------
        # Collect predictions across the validation set
        val_predictions_coco = get_model_predictions_in_coco_format(model, valid_dataloader, device, score_threshold=0.05)
        
        # Evaluate via pycocotools
        coco_metrics = coco_evaluate(val_coco_gt_dict, val_predictions_coco)
        
        # Print or log results
        print(f"Epoch [{epoch+1}/{epochs}] - Train Loss: {train_loss:.4f}, Valid Loss: {valid_loss:.4f}")
        print("COCO metrics:", coco_metrics)

        # Save checkpoint if best
        if valid_loss < best_loss:
            best_loss = valid_loss
            torch.save(model.state_dict(), checkpoint_path)
            training_metadata = {
                'epoch': epoch,
                'train_loss': train_loss,
                'valid_loss': valid_loss, 
                'learning_rate': lr_scheduler.get_last_lr()[0],
                'model_architecture': model.name,
                'coco_metrics': coco_metrics
            }
            with open(Path(checkpoint_path.parent/'training_metadata.json'), 'w') as f:
                json.dump(training_metadata, f)

    if device.type != 'cpu':
        getattr(torch, device.type).empty_cache()
```
